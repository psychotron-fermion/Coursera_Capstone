## 1.2.Working with url:
- creating a string variable **url** containing the url of the Wikipedia table;
- opening URL with the help of urllib.request and adding the HTML data into the **page** variable;
- parsing the HTML from the URL using the BeautifulSoup parse tree format;
- checking the structure of the HTML code with the Beautiful Soup’s prettify function.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "html")
print(soup.prettify())
```

## 1.3.Transforming data into a pandas' dataframe:
- retrieving all instances of the table tag present on the page;
- saving the instances in the **all_tables** array;
- bringing the data for the ”wikitable sortable” table;
- saving the data in the **”right_table“** variable.

```python
all_tables = soup.find_all("table")
print(all_tables)
```

```python
right_table = soup.find('table', class_='wikitable sortable')
print(right_table)
```

- assigning columns in the dataframe as three empty lists A, B, C;
- lopping through the rows in the table with the help of the Beautiful Soup function (rows start with the "tr" tag whereas data itself is located within the "td" tags);
- adding the rows to the variables **cells**;
- using **if condition** in order to check whether each **cells** array has three items;
- if the condition is true, the data is extracted and added into the array:

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

- creating a dataframe where each of the columns contains the data from the lists A, B and C respectively:

```python
df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighborhood']=C
print(df)
```

- filtering the dataframe dropping all the rows in which Borough has the value "Not assigned" and checking the shape of the new dataframe:

```python
newdf = df[~df["Borough"].str.contains("Not assigned")]
newdf.head()
```

### 2.3 Declaring a function for finding the coordinates of the neighbourhoods in Toronto:

```python
 def findCoord(PostalCode):   
    try: 
        geolocator = Nominatim(user_agent="toronto_explorer") 
        return geolocator.geocode(PostalCode) 
    
    except GeocoderTimedOut:  
        return findCoord(PostalCode)  
```

### 2.4 Using the postal codes of the neighbourhoods in Toronto from the **newdf** with the **findCoord** function and adding the values returned to the lists created above (**NAN** is added if there is no value returned):

```python
for i in (newdf["PostalCode"]): 
    if findCoord(i) != None: 
        loc = findCoord(i) 
        latitude.append(loc.latitude) 
        longitude.append(loc.longitude) 
    else: 
        latitude.append(np.nan) 
        longitude.append(np.nan) 
```

- Creating a method for getting the venues located in all the neighbourhoods contained in the **final_df**:

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## 3.4 Analysing each neighbourhood

- Creating a dataframe of all the neibourhoods and venue categories with the help of one-hot encoding:

```python
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

toronto_onehot = toronto_onehot.reindex(sorted(toronto_onehot.columns), axis=1)
```

```python
tor_df = pd.DataFrame(toronto_onehot)
columns = tor_df.columns
print(toronto_onehot.columns.get_loc("Neighborhood"))
```

```python
fixed_columns = [toronto_onehot.columns[193]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
toronto_onehot.head()
```

```python
toronto_onehot = toronto_onehot.loc[:,~toronto_onehot.columns.duplicated()]
toronto_onehot.head()
```

```python
toronto_onehot.shape
```

- Estimating the mean frequency of the occurrence of the venues of each category in the rows groupped by neighbourhood:

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
toronto_grouped.shape
```

- Printing each neighbourhood with five most frequent venues:

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

- Creating pandas' dataframe with the information about the neighbourhoods and the most frequently occurring venues:

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## 3.5 Clustering neighbourhoods in Toronto

- k-means algorithm is used for clustering neighbourhoods in Toronto (creates 5 clusters):

```python
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

kmeans.labels_[0:10] 
```

- adding the information about clusters to the **final_df** dataframe and merging it with all the related data:

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = final_df
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head()
```

- the **Cluster Labels** column has been coverted automatically into floats which makes it impossible to use its data for mapping. In order to a void this problem, cleaning the data in the column and its transformation into integer values was performed:

```python
toronto_merged = toronto_merged[~toronto_merged['Cluster Labels'].isnull()]
```

```python
toronto_merged['Cluster Labels'] = toronto_merged['Cluster Labels'].apply(np.int64)
```

```python
toronto_merged.head()
```

- mapping neighbourhoods divided into clusters:

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
